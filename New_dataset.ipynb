```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = "median")
imputer.fit(df)
```

```python
df.info()
```

```python
df_trans = imputer.transform(df)
df_trans
```

```python
df = pd.DataFrame(df_trans,columns = list(df.columns))
df.info()
```
